```python
import pandas as pd

data = pd.read_csv('Traffic_Crashes_-_Crashes.csv')
locations = data[['LATITUDE','LONGITUDE']]
locations = locations.dropna()
locations = locations[locations.LATITUDE != 0] #this was the issue plaguing the plot smh
```

Hello World

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import numpy as np

# Building the model 
kmeans = KMeans(n_clusters = 10) 
 
# Training the model 
kmeans.fit(locations) 
 
# Storing the predicted Clustering labels 
labels = kmeans.predict(locations) 
centers = np.array(kmeans.cluster_centers_)

plt.scatter(locations['LONGITUDE'], locations['LATITUDE'], c = labels,)
plt.scatter(centers[:,1], centers[:,0], marker='x', color='k', s=150)
plt.title("Center points of crashes in Chicago determined by K-Means")
plt.xlabel('LATITUDE')
plt.ylabel('LONGITUDE')
plt.show()
```

The X's in the graph are the cluster centers that were generated by the K-Means algorithm. The X's mean that those points in the clusters are where the crashes are the densest (the most crashes happen on or near the X in that cluster)

```python
kmeans2 = KMeans(n_clusters = 15) 
 
kmeans2.fit(locations) 
labels2 = kmeans2.predict(locations) 
centers2 = np.array(kmeans2.cluster_centers_)

plt.scatter(locations['LONGITUDE'], locations['LATITUDE'], c = labels2,)
plt.scatter(centers2[:,1], centers2[:,0], marker='x', color='k', s=150)
plt.title("Center points of crashes in Chicago determined by K-Means")
plt.xlabel('LATITUDE')
plt.ylabel('LONGITUDE')
plt.show()
```

```python
kmeans3 = KMeans(n_clusters = 20) 
 
kmeans3.fit(locations) 
labels3 = kmeans3.predict(locations) 
centers3 = np.array(kmeans3.cluster_centers_)

plt.scatter(locations['LONGITUDE'], locations['LATITUDE'], c = labels3,)
plt.scatter(centers3[:,1], centers3[:,0], marker='x', color='k', s=150)
plt.title("Center points of crashes in Chicago determined by K-Means")
plt.xlabel('LATITUDE')
plt.ylabel('LONGITUDE')
plt.show()
```

Interestingly, the clusters at the bottom of the graph don't change much between 15 and 20 clusters but the clusters at the top increase. This must mean that there are more crashes in north chicago.
