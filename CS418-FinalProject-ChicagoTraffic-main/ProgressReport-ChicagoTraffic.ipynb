**Crashes HeatMap**

<u>Hypothesis:</u> The highest recorded crash density would be in the Chicago Loop.

We wanted to test this hypothesis because we wanted to see how large of an effect does population density and the amount of vehicles in a specific area have on the number of crashes. Moreover, this is the area where most of our group members reside, so a natural curosity was also a driving force.

In order to plot this heat map we had to bucket the data which was provided in geographic coordinates in the dataset. We converted the longitude and latitude into multipoligon geometric object in python. Then we graphed these object on a 2-D plot. Then using a Chicago's map in a geojson format we further bucket our geometric objects into appropriate neighborhoods so that we can merge the dataframe and the geojson to plot. Then we finally plotted the merged data using geopandas. We used color as our channel to showcase the frequency of crashes on our map. 

After looking at the aggriated heatmap we realized that our hypothesis was not held true because there were relatively low crashes reported in the loop area. The intensity of crashes in the North of Chicago was more prominent. We do realize that this holds ground because North Chicago is more of a residential district and frequency of personal vehicles is more there. Whereas, the loop area has more public transportation traffic. Further demonstration of the results are mentioned in our machine learning analysis.



**<u>Machine Learning Analysis<u>**

One ML analysis we did was using K-Means on crash locations (latitude and longitude) to cluster them and determine the center (or most dense part) of each cluster. After cleaning the data we fit the model and then extracted the labels and cluster centers. Then we used those when plotting the locations to see the different clusters and the centers of the clusters on the scatter plot. We ran K-means 3 times with k = 10, 15, and 20. One thing we noticed was that the number of clusters in the bottom half of the plot didn’t increase between k = 15 and 20 and the number of clusters in the top half of the plot did increase between k = 15 and 20. This must mean that there are more crashes or the crashes are denser in the northern half of Chicago than the southern half. 

When comparing this to the heat map visualization, some of the cluster center points given in the k=10 plot correspond to neighborhoods that have higher numbers of crashes in the heat map. This means that k-means can be used to make initial predictions as to where the most crashes in an area occur, but a heat map is obviously more accurate in this matter. This is just based off of a side by side comparison. We can explore this comparison further by figuring out how to plot these cluster center points onto the heat map to get a more accurate sense of how the two plots correspond to each other. From there we can use the cluster centers from different k values to see what k value would give the most cluster centers in areas with more crashes to see if k-means is truly a good way to determine dense areas of crashes.
